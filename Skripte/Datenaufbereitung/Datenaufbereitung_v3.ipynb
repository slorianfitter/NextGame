```python
import pandas as pd
```

```python
reviews_data = pd.read_csv("../data/review_alltime.csv")
game_data = pd.read_csv("../data/wanted_information.csv")
```

```python
len(game_data)
```

```python
game_data.head()
```

```python
game_data["categories"] = game_data["categories"].fillna("NA")
game_data["categories"] = game_data["categories"].str.split(";")
```

```python
game_data["categories"] = game_data["categories"].apply(lambda x: [c.strip() for c in x])
game_data["categories"] = game_data["categories"].apply(lambda x: ";".join([c.strip() for c in x]))
```

```python
game_data["categories"].head()
```

```python
merged_data = pd.merge(game_data, reviews_data, on="id", how="left")
```

```python
merged_data["price"] = merged_data["price_in_cents_no_discount"].combine_first(merged_data["website_price"])
```

```python
final = merged_data.drop(labels=["Unnamed: 0", "total_steam_reviews_with_game", "website_price", "price_in_cents_no_discount"], axis=1)
```

```python
final = final[final["price"].notna()]
final = final.drop_duplicates(subset=["id"])
len(final)
```

```python
final.to_csv("../data/full_data_v1.csv", index = False)
```
